# Image Basics: Test 12

In this notebook, we will explore some fundamental operations on images using Python libraries such as OpenCV and Matplotlib. We will cover tasks from importing libraries to manipulating and displaying images.

```python
# Task 1
# Import the required libraries and confirm the imports

import cv2
import matplotlib.pyplot as plt

print("Libraries imported successfully")
```

## Task 2

Next, let's establish a path to our image file. We'll utilize an image file that's already available locally.

```python
# Task 2
# Define the filename for the local image

filename = "sample.jpg"
print("Image file name set to:", filename)
```

## Task 3

We'll read the image using OpenCV's `imread` function. OpenCV reads images in BGR format by default.

```python
# Task 3
# Read the image using cv2.imread

image_bgr = cv2.imread(filename)

# Verify that the image has been successfully loaded
if image_bgr is not None:
    print("Image loaded successfully.")
else:
    raise ValueError("Image could not be loaded.")
```

## Task 4

Let's check the type and shape of the loaded image to understand its dimensions and structure.

```python
# Task 4
# Print the type of the image
# Print the shape of the image

print("Type of image_bgr:", type(image_bgr))
print("Shape of image_bgr:", image_bgr.shape)
```

## Task 5

Convert the image from BGR to RGB format for proper display with Matplotlib (which expects RGB format).

```python
# Task 5
# Convert image_bgr to RGB

image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
print("Image converted to RGB format.")
```

## Task 6

Now, let's display the RGB image using Matplotlib.

```python
# Task 6
# Display the RGB image

plt.imshow(image_rgb)
plt.axis('off')  # Turn off the axes
plt.title('RGB Image')
plt.show()
```

## Task 7

Finally, let's extract the Red, Green, and Blue channels from the image and display each of them separately.

```python
# Task 7
# Extract and display each color channel

# Extract channels
red_channel = image_rgb[:, :, 0]
green_channel = image_rgb[:, :, 1]
blue_channel = image_rgb[:, :, 2]

# Display channels
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(red_channel, cmap='Reds')
axes[0].set_title('Red Channel')
axes[0].axis('off')

axes[1].imshow(green_channel, cmap='Greens')
axes[1].set_title('Green Channel')
axes[1].axis('off')

axes[2].imshow(blue_channel, cmap='Blues')
axes[2].set_title('Blue Channel')
axes[2].axis('off')

plt.show()
```
